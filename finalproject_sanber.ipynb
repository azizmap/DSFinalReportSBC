# Handling Outliers

outlier data replace with upperbound or lowerbound

```python
def outliers(df, column):
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3-q1
  lower_bound = q1-1.5*iqr
  upper_bound = q3+1.5*iqr

  for i in range(len(df)):
      if df[column].iloc[i] > upper_bound:
          df[column].iloc[i] = upper_bound
      if df[column].iloc[i] < lower_bound:
          df[column].iloc[i] = lower_bound
```

```python
columns = list(df.columns)
columns = columns[1:]

for column in columns:
  outliers(df, column)
```

# Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_train = df[columns]
data_train[columns] = scaler.fit_transform(data_train)
```

```python
data_train.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=100).fit(data_train)
labels = kmeans.labels_
```

```python
data_negara_label = df
data_negara_label["Cluster"] = labels
```

```python
data_clustering = data_train
data_clustering['Negara'] = df["Negara"]
data_clustering["Cluster"] = labels
```

```python
data_clustering
```

```python
cluster_map = {
    0 : "Layak Menerima Bantuan", 
    1 : "Tidak Layak Menerima Bantuan", 
    2 : "Kurang Layak Menerima Bantuan"
}

data_clustering = data_clustering.replace({"Cluster" : cluster_map})
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=data_clustering, x="GDPperkapita", y="Kematian_anak", hue="Cluster")
plt.suptitle('Clustering', y=0.95, weight="bold")
plt.title("semakin kecil GDP perkapita, semakin tinggi angka kematian anak", weight='light')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,8], centers[:, 0], c='Red', s = 200)

plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=data_clustering, x="GDPperkapita", y="Harapan_hidup", hue="Cluster")
plt.suptitle('Clustering', y=0.95, weight="bold")
plt.title("semakin kecil GDP perkapita, semakin rendah harapan hidup di negara tersebut", weight='light')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,8], centers[:, 6], c='Red', s = 200)

plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=data_clustering, y="Harapan_hidup", x="Kematian_anak", hue="Cluster")
plt.suptitle('Clustering', y=0.95, weight="bold")
plt.title("semakin rendah harapan hidup, semakin tinggi angka kematian anak di negara tersebut", weight='light')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:, 6], c='Red', s = 200)

plt.show()
```

```python
from tqdm import tqdm
#membuat array dengan nama wcss
wcss = []
#melakukan perulangan dari 1 sampai 11
for i in tqdm(range(1, 11)):
    #membuat nama variabel = kmeans yang mendeskripsikan jumlah cluster, perulangan k-means, dan random statenya
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 40)
    kmeans.fit(data_train)
    wcss.append(kmeans.inertia_)
#menampilkan plot wcss

plt.figure(figsize=(12,8))
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss)
#memberikan label plot dengan judul The Elbow Method
plt.title('The Elbow Method')
#memberikan label x dengan nama Number of clusters
plt.xlabel('Number of clusters')
#memberikan label y dengan nama WCSS
plt.ylabel('WCSS')
#menampilkan plot
plt.show()
```
